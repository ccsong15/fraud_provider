```python
import pandas as pd
```

```python
inpatient = pd.read_csv('/Users/cuicuisong/Desktop/data/inpatient.csv')
outpatient = pd.read_csv('/Users/cuicuisong/Desktop/data/outpatient.csv')
beneficiary = pd.read_csv('/Users/cuicuisong/Desktop/data/beneficiary.csv')
provider = pd.read_csv('/Users/cuicuisong/Desktop/data/provider.csv')
```

```python
inpatient.head()
```

```python
outpatient.head()
```

```python
# Merge inpatient and outpatient dataframes based on common columns
common_columns = [ idx for idx in outpatient.columns if idx in inpatient.columns]
print(common_columns)
Inpatient_Outpatient_Merge = pd.merge(inpatient, outpatient, left_on = common_columns, right_on = common_columns,how = 'outer')
```

```python
# Merge beneficiary details with inpatient and outpatient data
Inpatient_Outpatient_Beneficiary_Merge = pd.merge(Inpatient_Outpatient_Merge, beneficiary,
                                                  left_on='BeneID',right_on='BeneID',how='inner')
```

```python
# Merge provider details
Final_Dataset_Train = pd.merge(Inpatient_Outpatient_Beneficiary_Merge, provider , how = 'inner', on = 'Provider' )
```
